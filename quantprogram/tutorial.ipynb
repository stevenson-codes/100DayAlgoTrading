```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Download Ticker Data
aapl = yf.download('AAPL', start='2015-01-01', end='2025-01-01')
aapl
```

```python
# Download Multiple Ticker Data
tickers = ['AAPL', 'SPY', 'KO']
stocks = yf.download(tickers, start='2015-01-01', end='2025-01-01')
stocks
```
